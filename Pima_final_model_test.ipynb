```python
import numpy as np
import joblib
loaded_model = joblib.load('/content/drive/MyDrive/Raw CSV/diabetes_model.pkl')
loaded_scaler = joblib.load('/content/drive/MyDrive/Raw CSV/scaler.pkl')
sample = np.array([[6,148, 72, 35,0,33.6,0.627,50]])
sample2 = np.array([[8, 183, 80, 32, 210, 37.1, 0.480, 55]])
# Non-diabetic — young woman, normal glucose, healthy BMI, low family history
sample3 = np.array([[1, 99, 68, 24, 70, 27.3, 0.186, 28]])
# Correctly stack the samples vertically to create a 2D array for transformation
samples_to_transform = np.vstack([sample , sample2, sample3])
sample_scalled = loaded_scaler.transform(samples_to_transform)
pred = loaded_model.predict(sample_scalled)
prob = loaded_model.predict_proba(sample_scalled)

# Print results for each sample
for i in range(len(pred)):
    print(f"\nSample {i+1} Result:")
    print("  Diabetic" if pred[i]==1 else "  Not Diabetic")
    print("  Probability:", round(prob[i][1]*100, 1), "%")
```
